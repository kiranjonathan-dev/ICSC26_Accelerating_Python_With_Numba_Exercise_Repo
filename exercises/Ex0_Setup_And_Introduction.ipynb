## Using Timeit Line Magics

For the first two work books, we will mostly be using `timeit` line magics. There are a few ways we can use them.

```python
# If you haven't run the other cells already
import numpy as np
import matplotlib.pyplot as plt
```

### On A Single Function/Line

```python
# Single line
%timeit np.sum(np.arange(5))
```

```python
def sum_ints(n):
    return np.sum(np.arange(n))
```

```python
# Single function           
%timeit sum_ints(5)
```

```python
# With a custom number of loops per run
%timeit -n 100_000 sum_ints(5)
```

**Be careful with custom numbers of loops!** 

If very quick functions aren't looped enough times, the timings will be very noisy/inaccurate. `%timeit` generally does a good job of determining its own loop counts.

**Warning: You might wait longer for faster functions**

It might seem strange but quicker functions usually take longer to time, since they need to be run more times to get an accurate measurement. I know it makes them feel slow, but trust the data!

### On A Whole Cell

```python
%%timeit
x = np.arange(5)
np.sum(x)
```

### Storing The Timing Data

You can also store the timing data into a variable with `time = %timeit -o SOME_CODE`, which is perfect for storing values into an array for plotting

```python
time = %timeit -o np.sum(np.arange(5))

print(time.best)
print(time.worst)
print(time.average)
print(time.stdev)
print(time.all_runs)
# There are more options, you can use tab autocomplete to see what they are!
```

```python
problem_sizes = [10, 100, 1_000, 10_000]
times = []

for n in problem_sizes:
    tmp = %timeit -o np.sum(np.arange(n))
    times.append(tmp)

best_times = [time.best for time in times]

plt.plot(problem_sizes, best_times)
plt.xlabel("Problem Size")
plt.ylabel("Best Time")
plt.title("Timings for np.sum(np.arange(n))")
```
